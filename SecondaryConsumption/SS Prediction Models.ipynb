```python
SS_Pre = SS_Pre.drop(['preciptype'], axis=1)
SS_Post = SS_Post.drop(['preciptype'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
scaler = StandardScaler()
SS_Pre_t = scaler.fit_transform(SS_Pre)
SS_Post_t = scaler.transform(SS_Post)
```

```python
def create_sequences(x, y, time_steps=7):  
    x_seq, y_seq = [], []
    for i in range(len(x) - time_steps):
        x_seq.append(x[i:i+time_steps])
        y_seq.append(y[i+time_steps])
    return np.array(x_seq), np.array(y_seq)

x_train, y_train = create_sequences(SS_Pre_t, SS_Pre_t, 7) 
x_test, y_test = create_sequences(SS_Post_t, SS_Post_t, 7)  

y_train = y_train.reshape(y_train.shape[0], 1, y_train.shape[1])
y_test = y_test.reshape(y_test.shape[0], 1, y_test.shape[1])
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

### Linear Regression

```python
lasso_model = Lasso(alpha=0.1)  # You can adjust the alpha parameter to change the amount of regularization

lasso_model.fit(x_train.reshape(x_train.shape[0], -1), y_train.reshape(y_train.shape[0], -1))

y_pred_lasso = lasso_model.predict(x_test.reshape(x_test.shape[0], -1))

mse_lasso = mean_squared_error(y_test.reshape(y_test.shape[0], -1), y_pred_lasso)
print("Lasso Regression MSE:", mse_lasso)
```

### Decision Tree

```python
dt_model = DecisionTreeRegressor()
dt_model.fit(x_train.reshape(x_train.shape[0], -1), y_train.reshape(y_train.shape[0], -1))
y_pred_dt = dt_model.predict(x_test.reshape(x_test.shape[0], -1))
mse_dt = mean_squared_error(y_test.reshape(y_test.shape[0], -1), y_pred_dt)
print("Decision Tree MSE:", mse_dt)
```

### Random Forest

```python
rf_model = RandomForestRegressor()
rf_model.fit(x_train.reshape(x_train.shape[0], -1), y_train.reshape(y_train.shape[0], -1))
y_pred_rf = rf_model.predict(x_test.reshape(x_test.shape[0], -1))
mse_rf = mean_squared_error(y_test.reshape(y_test.shape[0], -1), y_pred_rf)
print("Random Forest MSE:", mse_rf)
```

### LSTM

```python
import tensorflow as tf
def lstmModel():
    Input = tf.keras.layers.Input(shape=(7,x_train.shape[2]))
    x = tf.keras.layers.LSTM(256,return_sequences=True)(Input)
    x = tf.keras.layers.LSTM(256)(x)
    x = tf.keras.layers.Dense(256,activation=tf.nn.leaky_relu)(x)
    x = tf.keras.layers.Dense(y_train.shape[2])(x)
    model = tf.keras.Model(inputs=Input,outputs=x)
    return model
```

```python
lstm_Model = lstmModel()
lstm_Model.summary()
```

```python
lstm_Model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss=tf.keras.losses.MeanSquaredError(), metrics=[tf.keras.metrics.MeanSquaredError()])
```

```python
ckpt = tf.keras.callbacks.ModelCheckpoint(filepath='lstm_model.pb',save_weights_only=True,save_best_only=True,monitor='val_mean_squared_error',mode='min')
```

```python
lstm_Model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100, batch_size=32, shuffle=False, callbacks=[ckpt])
```
